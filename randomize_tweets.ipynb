```python
import pandas as pd
import glob
import os
from collections import defaultdict
import numpy as np
import locationtagger
from pathlib import Path

USERS_DATA = r"D:/coding/classes/csci461/Rhino-Sentiment-Analysis-Project/processed_users/"
```

```python
#take in user data
path = USERS_DATA
print(path)
all_files = glob.glob(os.path.join(path , "*.csv"))
print(os.path.join(path , "*.csv"))
li = []

for filename in all_files:
    print(filename)
    df = pd.read_csv(filename, index_col=None, header=0, lineterminator='\n', converters = {'id': str})
    li.append(df)


users = pd.concat(li, axis=0, ignore_index=True)
users.rename(columns={'created_at': 'users_created_at'}, inplace=True)
```

```python
#take in tweet
path = r'D:/coding/classes/csci461/Rhino-Sentiment-Analysis-Project/filtered_tweets' # use your path
print(path)
all_files = glob.glob(os.path.join(path , "*.csv"))
print(os.path.join(path , "*.csv"))
li = []

for filename in all_files:
    print(filename)
    df = pd.read_csv(filename, index_col=None, header=0, lineterminator='\n', converters = {'author_id': str})
    li.append(df)


tweets = pd.concat(li, axis=0, ignore_index=True)
tweets.rename(columns={'created_at': 'tweet_created_at'}, inplace=True)
tweets.drop( columns="created_at\r" ,inplace=True)
```

```python
users = users.set_index("id")
```

```python
tweet_w_users = tweets.join(users, 
           lsuffix="_users", 
           rsuffix="_tweets", 
           on="author_id",
            how="inner" )
```

```python
us_only_tweets = tweet_w_users[tweet_w_users["country"] == "United States"]
```

```python
us_only_tweets = us_only_tweets.drop(["Unnamed: 0", "created_at\r\r", "city\r"], axis=1)
```
